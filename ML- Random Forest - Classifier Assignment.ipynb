```python
import pandas as pd
```

```python
dataset=pd.read_csv('CKD.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset['classification_yes'].value_counts()
```

```python
indep=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dep=dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(indep, dep, random_state=0, test_size=0.30)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
grid_params={
    'n_estimators':[100,20,300,400,500],
    'criterion':['gini', 'entropy', 'log_loss'],
    'max_features':['sqrt', 'log2']
}
classifier=GridSearchCV(RandomForestClassifier(),grid_params, verbose=3, n_jobs=-1, refit=True, scoring='f1')
classifier.fit(X_train, Y_train)
```

```python
re=classifier.cv_results_
```

```python
print(re)
```

```python
Y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test, Y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
clf=classification_report(Y_test, Y_pred)
print(clf)
```
